## ðŸ–¥ï¸ SERVER CODE (Server-PT BaseStation)

```python
# SERVER - Data Collection Center
# IP: 192.168.1.1, Port: 5000
# Receives aggregated data from all Cluster Heads

from udp import *
from time import *

# Statistics tracking
total_packets = 0
cluster_stats = {
    "CH1": 0,
    "CH2": 0,
    "CH3": 0
}

def onUDPReceive(ip, port, data):
    global total_packets, cluster_stats
    
    total_packets = total_packets + 1
    message = str(data)
    
    # Extract cluster ID
    if "CH1" in message:
        cluster_stats["CH1"] = cluster_stats["CH1"] + 1
        cluster_name = "CLUSTER 1"
    elif "CH2" in message:
        cluster_stats["CH2"] = cluster_stats["CH2"] + 1
        cluster_name = "CLUSTER 2"
    elif "CH3" in message:
        cluster_stats["CH3"] = cluster_stats["CH3"] + 1
        cluster_name = "CLUSTER 3"
    else:
        cluster_name = "UNKNOWN"
    
    print("=" * 50)
    print("SERVER RECEIVED DATA!")
    print("From: " + ip + ":" + str(port) + " [" + cluster_name + "]")
    print("Data: " + message)
    print("Total Packets: " + str(total_packets))
    print("CH1: " + str(cluster_stats["CH1"]) + " | CH2: " + str(cluster_stats["CH2"]) + " | CH3: " + str(cluster_stats["CH3"]))
    print("=" * 50)

def main():
    socket = UDPSocket()
    socket.onReceive(onUDPReceive)
    socket.begin(5000)
    
    print("*" * 50)
    print("     MULTI-CLUSTER SERVER STARTED")
    print("*" * 50)
    print("Listening on: 192.168.1.1:5000")
    print("Expecting data from 3 Cluster Heads")
    print("Waiting for data...")
    print("")
    
    counter = 0
    while True:
        counter = counter + 1
        if counter % 30 == 0:
            print("[Server Heartbeat] Running... (" + str(counter / 30) + " min)")
        sleep(1)

if __name__ == "__main__":
    main()
```

## ðŸŽ¯ CLUSTER HEAD 1 CODE (ClusterHead-1)

```python
# CLUSTER HEAD 1
# IP: 192.168.1.100, Listen Port: 5001
# Receives from Member1-1 and Member1-2, forwards to Server

from udp import *
from time import *

# Configuration
CLUSTER_ID = "CH1"
LISTEN_PORT = 5001
SERVER_IP = "192.168.1.1"
SERVER_PORT = 5000

# Data buffer
data_buffer = {}
cycle_count = 0

def onUDPReceive(ip, port, data):
    global data_buffer, cycle_count
    
    message = str(data)
    print("Received from " + ip + ": " + message)
    
    # Parse message format "M1-1:0" or "M1-2:1"
    if ":" in message:
        parts = message.split(":")
        node_id = parts[0]
        sensor_val = parts[1]
        
        # Store in buffer
        data_buffer[node_id] = sensor_val
        
        # Check if received from both members
        if "M1-1" in data_buffer and "M1-2" in data_buffer:
            # Create aggregated message
            agg_msg = CLUSTER_ID + "|M1-1:" + data_buffer["M1-1"] + "|M1-2:" + data_buffer["M1-2"]
            
            # Forward to server
            socket.send(SERVER_IP, SERVER_PORT, agg_msg)
            
            cycle_count = cycle_count + 1
            print(">>> Forwarded to SERVER - Cycle " + str(cycle_count))
            print(">>> Data: " + agg_msg)
            print("")
            
            # Clear buffer
            data_buffer = {}

def main():
    global socket
    socket = UDPSocket()
    socket.onReceive(onUDPReceive)
    
    result = socket.begin(LISTEN_PORT)
    print("=" * 40)
    print("CLUSTER HEAD 1 STARTED")
    print("=" * 40)
    print("Cluster ID: " + CLUSTER_ID)
    print("Listening on port: " + str(LISTEN_PORT))
    print("Server: " + SERVER_IP + ":" + str(SERVER_PORT))
    print("Socket status: " + str(result))
    print("Waiting for Member1-1 and Member1-2...")
    print("")
    
    counter = 0
    while True:
        counter = counter + 1
        if counter % 20 == 0:
            print("[CH1 Status] Active... (" + str(counter / 20) + ")")
        sleep(1)

if __name__ == "__main__":
    main()
```

### Member1-1 (Cluster 1, Node 1)

```python
# MEMBER 1-1 (Cluster 1, Node 1)
# IP: 192.168.1.101
# Sends to ClusterHead-1 (192.168.1.100:5001)

from udp import *
from gpio import *
from time import *

# Configuration
NODE_ID = "M1-1"
CH_IP = "192.168.1.100"
CH_PORT = 5001
SENSOR_PIN = 0

# Setup
pinMode(SENSOR_PIN, INPUT)

def main():
    socket = UDPSocket()
    socket.begin(5011)
    
    print("Member 1-1 Started [Cluster 1]")
    print("Node: " + NODE_ID)
    print("Target CH: " + CH_IP + ":" + str(CH_PORT))
    
    packet_count = 0
    
    while True:
        sensor_value = digitalRead(SENSOR_PIN)
        message = NODE_ID + ":" + str(sensor_value)
        
        socket.send(CH_IP, CH_PORT, message)
        print("Packet " + str(packet_count) + " -> " + message)
        
        packet_count = packet_count + 1
        sleep(2)

if __name__ == "__main__":
    main()
```
